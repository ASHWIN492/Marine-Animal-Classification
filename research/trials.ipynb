```python
import cv2
import numpy as np

image_path = "C:/Users/ashwi/OneDrive/Desktop/54403656_01419609db_b.jpg"
input_size = (300,225)

# Try to load the image
img = cv2.imread(image_path)

# Check if the image is loaded successfully
if img is not None:
    img = cv2.resize(img, input_size)
    img = img / 255.0
    img = np.expand_dims(img, axis=0)
    print("Image successfully loaded and preprocessed.")
else:
    print("Error loading the image. Please check the image path.")
```

```python
from tensorflow.keras.models import load_model

model_path = 'C:/Users/ashwi/OneDrive/Desktop/devrev07_1/under-water-animal-classification/artifacts/training/model.h5'
model = load_model(model_path)
```

```python
predictions = model.predict(img)
```

```python
class_labels = ['Clams', 'Corals', 'Crabs','Dolphin','Eel','Fish','Jelly Fish','Jelly Fish','Lobster','Nudibranchs','Octopus','Otter','Penguin','Puffers','Sea Rays','Sea Urchins','Seahorse','Seal','Sharks','Shrimp','Squid','Starfish','Turtle_Tortoise','Whale']  
print(predictions.sort()[0)
predicted_class = np.argmax(predictions)
predicted_label = class_labels[predicted_class]

print(f'Predicted Class: {predicted_class} ({predicted_label})')
```

```python
import matplotlib.pyplot as plt

plt.imshow(cv2.cvtColor(cv2.imread(image_path), cv2.COLOR_BGR2RGB))
plt.title(f'Predicted: {predicted_label}')
plt.show()
```
